```python
def decimal_to_binary(decimal_num):
    binary_list = []
    while decimal_num > 0:
        remainder = decimal_num % 2
        binary_list.insert(0, remainder)
        decimal_num = decimal_num // 2
    binary_str = ''.join(str(i) for i in binary_list)
    return binary_str
decimal_num = 89
binary_num = decimal_to_binary(decimal_num)
print("Binary equivalent of", decimal_num, "is", binary_num)
```

```python
numeros = [20,5,10,8]
print(*numeros)
resultados = []
while len(numeros) > 1:
    for i in range(len(numeros)-1):
        elemento1 = numeros[i]
        elemento2 = numeros[i+1]
        suma = elemento1 + elemento2
        resultados.append(suma)
    print(" " * (len(numeros)-2), end="")
    print(*resultados)
    numeros = resultados
    resultados = []
```
